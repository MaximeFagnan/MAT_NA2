```python
# EXÉCUTEZ CETTE CELLULE POUR VOIR DES RECTANGLES SVP!!!

import numpy as np
import matplotlib.pyplot as plt

# Fonction cible
def f(x):
    return x**2  # Exemple : x^2

# Graphique avec rectangles de Riemann (méthode par la droite)
def plot_rect_droite(a, b, n):
    x = np.linspace(a, b, 1000)  # Points pour tracer la courbe
    y = f(x)
    
    dx = (b - a) / n
    x_rect = np.linspace(a , b-dx, n)  # Points à gauche des intervalles
    y_rect = f(x_rect+dx)  # f(x_droite)
    
    # Tracé du graphique
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="f(x) = x²", color="blue")
    plt.bar(x_rect, y_rect, width=dx, align="edge", color="orange", edgecolor="black", alpha=0.6)
    
    plt.title(f"Rectangles de Riemann (droite) avec n = {n}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
```

On peut ainsi faire apparaitre des rectangles. La borne de gauche est `a`, la borne de droite est `b` et le nombre de rectangle est `n`.

```python
# Exemple simple
a = 0  # borne de gauche
b = 3  # borne de droite
n = 4  # nombre de rectangle
plot_rect_droite(a, b, n)
```

## Partie 2: L'intégrale définie

On dit qu'une fonction est intégrable (au sens de Riemann) si 
$$\lim\limits_{n\to\infty}A_n=\lim\limits_{n\to\infty} \sum_{i=1}^n f(a+\Delta x \cdot i)  \Delta x \quad \text{où }\; \Delta x \; \text{est définie comme plus haut}$$
converge.


**Définition** : L'**intégrale définie** d'une fonction $f$ de `a` à `b` est donnée par la limite suivante (si elle converge).
$$\int_a^b f(x) dx \stackrel{\text{def}}{=}\lim\limits_{n\to\infty} \sum_{i=1}^n f(a+\Delta x \cdot i) \Delta x$$
Cette formule donne l'aire nette sous la courbe $f$.

Vous allez calculer l'intégrale définie de $f(x)=x^2$ pour `a=0` jusqu'à un `b` quelconque dans la prochaine cellule, soit 
\begin{align*}
\int_0^b x^2 dx
\end{align*}

**Instruction**: Complétez les 3 dernières lignes.

\begin{align*}
\int_0^b f(x) dx &= \lim\limits_{n\to\infty} \sum_{i=1}^n f(0+\Delta x \cdot i) \Delta x \\
&= \lim\limits_{n\to\infty} \Delta x \sum_{i=1}^n  (\Delta x \cdot i)^2 \\
&= \lim\limits_{n\to\infty} \Delta x^3 \sum_{i=1}^n  i^2 \\
&= \lim\limits_{n\to\infty} \text{(Substituez $\Delta x$ et utilisez la formule de somme des carrés)} \\
&= \lim\limits_{n\to\infty} \text{(Simplifiez la forme $\frac{\infty}{\infty}$)} \\
&= \text{(résolution de la limite)}
\end{align*}
