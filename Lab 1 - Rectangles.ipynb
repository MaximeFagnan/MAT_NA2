Voici du code qui permet de visualiser des rectangles de gauche et des rectangles de droite.

```python
# EXÉCUTEZ CETTE CELLULE POUR VOIR DES RECTANGLES SVP!!!

import numpy as np
import matplotlib.pyplot as plt

# Fonction
def f(x):
    return x**2  # Exemple : x^2

fig = (6,2.5)

# Graphique avec rectangles de Riemann (méthode par la droite)
def plot_rect_droite(a, b, n):
    x = np.linspace(a, b, 1000)  # Points pour tracer la courbe
    y = f(x)

    dx = (b - a) / n
    x_rect = np.linspace(a , b-dx, n)  # Points à gauche des intervalles
    y_rect = f(x_rect+dx)  # f(x_droite)
    somme_des_aires = sum(y_rect*dx)

    # Tracé du graphique
    plt.figure(figsize=fig)
    plt.plot(x, y, label="f(x) = x²", color="blue")
    plt.bar(x_rect, y_rect, width=dx, align="edge", color="orange", edgecolor="black", alpha=0.6)

    plt.title(f"Rectangles de Riemann (droite) avec n = {n}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
    print(f"A_D = {somme_des_aires}")


# Graphique avec rectangles de Riemann (méthode par la gauche)
def plot_rect_gauche(a, b, n):
    x = np.linspace(a, b, 1000)  # Points pour tracer la courbe
    y = f(x)

    dx = (b - a) / n
    x_rect = np.linspace(a , b-dx, n)  # Points à gauche des intervalles
    y_rect = f(x_rect)  # f(x_gauche)
    somme_des_aires = sum(y_rect*dx)

    # Tracé du graphique
    plt.figure(figsize=fig)
    plt.plot(x, y, label="f(x) = x²", color="blue")
    plt.bar(x_rect, y_rect, width=dx, align="edge", color="orange", edgecolor="black", alpha=0.6)

    plt.title(f"Rectangles de Riemann (gauche) avec n = {n}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
```

On peut ainsi faire apparaitre des rectangles. La borne de gauche est `a`, la borne de droite est `b` et le nombre de rectangle est `n`.

```python
# Exemple simple
a = 0  # borne de gauche
b = 3  # borne de droite
n = 4  # nombre de rectangle
plot_rect_droite(a, b, n)
```

## Partie 2: L'intégrale définie

On dit qu'une fonction est intégrable (au sens de Riemann) de `a` à `b` si
$$A=\lim\limits_{n\to\infty}A_D=\lim\limits_{n\to\infty} \sum_{i=1}^n f(a+\Delta x \cdot i)  \Delta x \quad \text{où }\; \Delta x \; \text{est définie comme plus haut}$$
converge.

**Idée :** Pour toutes les belles fonctions que vous avez utilisez dans votre vie, la limite des rectangles de droite converge si et seulement si la limite des rectangles de gauche converge. De plus, si ces limitent convergent, ils convergent vers la même valeur $A$.
$$\lim\limits_{n\to\infty} \text{A}_D = \lim\limits_{n\to\infty} \Delta x \sum_{i=1}^n f(a+\Delta x \cdot i) = A = \lim\limits_{n\to\infty} \text{A}_G = \lim\limits_{n\to\infty} \Delta x \sum_{i=0}^{n-1} f(a+\Delta x \cdot i)$$

**Définition** : L'**intégrale définie*** d'une fonction $f$ de `a` à `b` est donnée par la limite suivante (si elle converge). Notez que c'est le même $\Delta x$ que vous avez définie plus haut.
$$\int_a^b f(x) dx \stackrel{\text{def}}{=}\lim\limits_{n\to\infty} \sum_{i=1}^n f(a+\Delta x \cdot i) \Delta x$$
Cette formule donne l'aire nette sous la courbe $f$.

*En fait presque, mais cette définition fonction pour toutes les belles fonctions que vous avez utilisez dans un cours de mathématique. Pour les fonctions pathologique et dégueulasse, il faut créer une définition plus pointue.
